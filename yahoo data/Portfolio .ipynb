# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

Leer datos en el directorio

```python
files = [file for file in os.listdir("./historicalData")] #lista con los archivos
```

```python
dataFrames = []

for file in files:
    df = pd.read_csv("./historicalData/" + file)
    df =  {file:df["Adj Close"]} 
    df = pd.DataFrame(df)
    dataFrames.append(df)
```

Concatenar datos axis = 1

```python
allData = pd.concat([dataframe for dataframe in dataFrames], axis=1)
allData.head()
```

```python
for dataframe in dataFrames:
    dataframe.plot(title="Historical Prices")
```

```python
allData.to_excel('portfolio.xlsx', index = False)#Exportar los datos del dataframe a un excel
```
